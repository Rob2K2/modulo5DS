# Ejemplo SPC (Statistical Process Control) 

## PySpc es una biblioteca de Python destinada a hacer que los Gráficos de control de procesos estadísticos sean lo más fáciles posible.
 



### Instalamos la libreria pyspc

```python
pip install pyspc
```

### Importamos las librerias necesarias

```python
import numpy
from pyspc import *
```

### El conjunto de datos de ejemplo son de la fabricación de anillos de piston donde se detallan las variaciones de muestra de las versiones del producto
### Creamos un dataset con el conjunto de datos de ejemplo

```python
pistonring_df = pd.read_csv('pistonrings.csv')
pistonring_df.head(3)
```

### Mostramos los gráficos de suma acumulada y promedio móvil ponderado

```python
a = spc(pistonring_df) + cusum() + ewma() + rules()
print(a)
```

### Mostramos los gráficos de medición de la estabilidad del proceso

```python
b = spc(pistonring_df) + xbar_rbar() + rbar() + rules()
print(b)
```
